```python
df2["average"]=(df2["Open"]+df2["High"]+df2["Low"]+df2["Close"])/4
```

```python
fig, ax = plt.subplots(figsize=(15, 9))
x = np.arange(len(df2["Date"]))
ax.plot(x, df2["average"], color='blue', linewidth=1, linestyle='-')
l, = ax.plot(x, df2["Low"], color='orange', linewidth=1)
l.set_linestyle('-');
```
