## matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2.setup plot
fig,ax=plt.subplots(figsize=(50,20))
# x-axis=50 , y-axis=20
                    
ax.plot(x,y)

# 4.Customize plot
ax.set(title="simple plt",xlabel="x-asix",ylabel="y-axis")

# 5.save & show(save the figure)
fig.savefig("images/sample-plot.png")
```

# which one should I use
* when something quickly, use pyplot method
* when plotting something more advanced ,use the OO method

```python
hd.head()
```

```python
over50=hd[hd["age"]>50]
over50.head()
```

```python
over50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,6))
```

```python
#OO method mixed with piplot method
fig,ax=plt.subplots(figsize=(10,6))
over50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
ax.set_xlim([45,100])
```

```python
## oo method from scratch
fig,ax=plt.subplots(figsize=(10,6))
## plot the data
scatter=ax.scatter(x=over50["age"],
                  y=over50["chol"],
                  c=over50["target"])
#customize the plot
ax.set(title="Heart Disease & Cholesterol Levels",xlabel="Age",ylabel="Cholesterol")

#add legend
ax.legend(*scatter.legend_elements(),title="Target")

##add a horizontal line
#ax.axhline(over50["chol"].mean())
ax.axhline(over50["chol"].mean(),linestyle="--")
```

```python
#subplot fo chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

scatter=ax0.scatter(x=over50["age"],y=over50["chol"],c=over50["target"])
ax0.set(title="Heart Disease and cholesterol lavels",xlabel="age",ylabel="cholesterol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(y=over50["chol"].mean(),linestyle="--")

scatter=ax1.scatter(x=over50["age"],y=over50["thalach"],c=over50["target"])
ax1.set(title="Heart Disease and Max heart rate",xlabel="age",ylabel="Max heart rate")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(y=over50["thalach"].mean(),linestyle="--")
```

```python
#share x
#subplot fo chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

scatter=ax0.scatter(x=over50["age"],y=over50["chol"],c=over50["target"])
ax0.set(title="Heart Disease and cholesterol lavels",xlabel="age",ylabel="cholesterol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(y=over50["chol"].mean(),linestyle="--")

scatter=ax1.scatter(x=over50["age"],y=over50["thalach"],c=over50["target"])
ax1.set(title="Heart Disease and Max heart rate",xlabel="age",ylabel="Max heart rate")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(y=over50["thalach"].mean(),linestyle="--")

#add a title to the figure 
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")
```

```python
plt.style.use("seaborn-whitegrid")

#subplot fo chol,age,thalach

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

scatter=ax0.scatter(x=over50["age"],y=over50["chol"],c=over50["target"],cmap="winter")#this change the color schame
ax0.set(title="Heart Disease and cholesterol lavels",xlabel="age",ylabel="cholesterol")

#change x axis limits in ax0
ax0.set_xlim([50,80])
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(y=over50["chol"].mean(),linestyle="--")

scatter=ax1.scatter(x=over50["age"],y=over50["thalach"],c=over50["target"],cmap="summer")
ax1.set(title="Heart Disease and Max heart rate",xlabel="age",ylabel="Max heart rate")

#change x axis limits in ax1
ax1.set_xlim([50,80])

#change y axis limits in ax1
ax1.set_ylim([60,200])
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(y=over50["thalach"].mean(),linestyle="--")
```
