```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor


```

```python
np.random.seed(42)
data=pd.read_csv("..\data/car-sales-extended-missing-data.csv")
data.dropna(subset=["Price"],inplace=True)

cat_impute=SimpleImputer(strategy="constant",fill_value="missing")
door_impute=SimpleImputer(strategy="constant",fill_value=4)
num_impute=SimpleImputer(strategy="mean")

c_f=["Make","Colour"]
d_f=["Doors"]
num_f=["Odometer (KM)"]

imputer=ColumnTransformer([("cat_impute",cat_impute,c_f),("num_impute",num_impute,num_f),("door_impute",door_impute,d_f)])

X=data.drop("Price",axis=1)
y=data["Price"]


filled_X=imputer.fit_transform(X)
X=pd.DataFrame(filled_X,columns=["Make","Colour","Odometer (KM)","Doors"])


one_hot=OneHotEncoder()
cf=["Make","Colour","Doors"]
transform_x=ColumnTransformer([("one_hot",one_hot,cf)],remainder="passthrough")
transform=transform_x.fit_transform(X)


X_train,X_test,y_train,y_test=train_test_split(transform,y,test_size=0.2)
model=RandomForestRegressor()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
