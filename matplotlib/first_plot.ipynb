# Making figures with Numpy arrays

```python
import numpy as np
```

```python
# create some data
x=np.linspace(0,10,100)
x
```

```python
# create a line plot
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
# create a scatter plot

fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig,ax=plt.subplots(figsize=(13,5))
ax.scatter(x,np.sin(x));
```

```python
# create a bar plot

nut_butter_prices={"almond buuter":10,"peanut butter":6,"Cashew butter":12}
fig,ax=plt.subplots(figsize=(5,3))
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Dan's nut Butter Store",xlabel="butter",ylabel="Price($)");
```

```python
#histograms
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(100));
```

## Which one should use??

* When plotting something quickly use pyplot method.
* when plotting something me advance use the oo method.

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50.plot(figsize=(10,5),kind="scatter",x="age",y="chol",c="target")
```

```python
# oo method

fig,axis=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=axis);
axis.set_xlim([40,100])
```

```python
# OO method form scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
ax.set(title="Heart disease and Cholesterol levels",xlabel="Age",ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.head()
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(title="Heart disease and cholesterol levels",xlabel="Age",ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(y=over_50["chol"].mean(),linestyle="--")


scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="Heart disease and thalach Level",xlabel="Age",ylabel="Thalach")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");


fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid")

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax0.set(title="Heart disease and cholesterol levels",ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.set_xlim([50,80])
ax0.axhline(y=over_50["chol"].mean(),linestyle="--")


scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="plasma")
ax1.set(title="Heart disease and thalach Level",xlabel="Age",ylabel="Thalach")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
ax1.set_xlim([50,80])

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
