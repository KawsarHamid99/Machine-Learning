### K Means Clustering Algrithm

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("income.csv")
dataset.head(3)
```

```python
dataset.tail()
```

```python
plt.scatter(dataset["Age"],dataset["Income($)"])
```

```python
model=KMeans(n_clusters=3)
```

```python
y_predict=model.fit_predict(dataset[["Age","Income($)"]])
dataset["Cluster"]=y_predict
```

```python
dataset
```

```python
df1=dataset[dataset.Cluster==0]
df2=dataset[dataset.Cluster==1]
df3=dataset[dataset.Cluster==2]

plt.scatter(df1.Age,df1["Income($)"],color="green")
plt.scatter(df2.Age,df2["Income($)"],color="red")
plt.scatter(df3.Age,df3["Income($)"],color="blue")

plt.xlabel("Age")
plt.ylabel("Income($)")
```

```python
scaler=MinMaxScaler()
scaler.fit(dataset[["Income($)"]])
```

```python
dataset["Income($)"]=scaler.transform(dataset[["Income($)"]])
```

```python
dataset
```

```python
scaler.fit(dataset[["Age"]])
dataset["Age"]=scaler.transform(dataset[["Age"]])
dataset
```

## 

```python
y_predict_sc=model.fit_predict(dataset[["Age","Income($)"]])
dataset["Cluster_sc"]=y_predict_sc
dataset
```

```python
dfs1=dataset[dataset.Cluster_sc==0]
dfs2=dataset[dataset.Cluster_sc==1]
dfs3=dataset[dataset.Cluster_sc==2]

plt.scatter(dfs1.Age,dfs1["Income($)"],color="green",label='Income($)')
plt.scatter(dfs2.Age,dfs2["Income($)"],color="red",label='Income($)')
plt.scatter(dfs3.Age,dfs3["Income($)"],color="blue",label='Income($)')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color="purple",marker="*",label='centroid')
plt.legend()

plt.xlabel("Age")
plt.ylabel("Income($)")
```

### Find The ELbow

```python
model.cluster_centers_
```

```python
sse=[]
for i in range(1,10):
    MODEL=KMeans(n_clusters=i)
    MODEL.fit(dataset[["Age","Income($)"]])
    sse.append(MODEL.inertia_)
```

```python
k_rng=range(1,10)
plt.xlabel("K")
plt.ylabel("sum of squred error")
plt.plot(k_rng,sse)
```
